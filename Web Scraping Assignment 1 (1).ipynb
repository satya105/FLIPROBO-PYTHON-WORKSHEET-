# Answer 4- Python program to display list of respected former presidents of India(i.e. Name , Term of office) 

```python
Indian_President_page=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
soup=BeautifulSoup(Indian_President_page.content)
```

```python
president=soup.find_all('div', class_= "presidentListing")
president_name=[]
tenure=[]
for i in president:
    var1 = i.h3.text.split("(")
    president_name.append(var1[0])
    tenure.append(i.p.text.strip("Term of Office: "))
```

```python
import pandas as pd
President_list=pd.DataFrame({})
President_list["Name"]=president_name
President_list["Tenure"]=tenure
President_list
```

# # Answer 7- Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world :
i) Headline
ii) Time
iii) News Link

```python
cnbc_page= requests.get("https://www.cnbc.com/world/?region=world")
soup= BeautifulSoup(cnbc_page.content)
headline=soup.find_all('div', class_="RiverPlusCard-cardLeft")
#print(headline)
for i in headline:
    print("################################")
    print("Headline",i.a.text)
    headtime=i.find_all('div',class_="RiverByline-authorBylineContainer")
    for j in headtime:
        print("Time",j.span.text)
    print("News_Link",i.a['href'])
    print("*******************************")
```

```python
movie_ratings=soup_1.find_all('td',class_='ratingColumn imdbRating')

column_ratings=[]
for i in movie_ratings[0:100]:
    column_ratings.append(i.text.strip())
#print(column_ratings[:100])
```

```python
import pandas as pd
TopMovies=pd.DataFrame({})
TopMovies['Names']=movies_100
TopMovies['ReleaseYear']=column_year
TopMovies['Rating']=column_ratings

TopMovies 
```

# Answer 9-  Python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL# 

```python
# ANSWER 9(i)-Restaurant name
dineout= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
Soup=BeautifulSoup(dineout.content)
#print(dineout)

restaurent_name= Soup.find_all('div',class_="restnt-info cursor")
#print(restaurent_name)
for i in restaurent_name:
    print("Restaurant Name :",i.a.text)

```

```python
# ANSWER 9(ii)- Cuisine
cuis_name= Soup.find_all('a',href="/delhi-restaurants/north-indian-cuisine")
for d in cuis_name:
    print("Cuisine :",d.text)
cuis_name2= Soup.find_all('a', href="/delhi-restaurants/chinese-cuisine")
for d in cuis_name2:
    print("Cuisine :",d.text)
```

```python
# ANSWER 9(iii)- Location
for i in restaurent_name:
    loc_names= i.find_all('div',class_="restnt-loc ellipsis")
    #print(loc_names)
    for l in loc_names:
        print("Location :",l.text)
```

```python
# ANSWER 9(iv)- Ratings
restaurent=Soup.find_all('div',class_="restnt-main-wrap clearfix")
for c in restaurent:
    user_rate=c.find_all('div',class_="restnt-rating rating-4")
    for d in user_rate:
        print("Ratings :",d.text)
```

```python
# ANSWER 9(v)-   IMAGE URL  
image_url= Soup.find_all('img',class_="no-img")

image_columns=[]
for i in image_url:
    image_columns.append(i['data-src'].strip('[]'))
print(image_columns)
```

# Answer 10- Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank
ii) Publication
iii) h5-index
 iv) h5-median

```python
google_scholar_form=requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
#print(google_scholar_form)
soup4=BeautifulSoup(google_scholar_form.content)
#print(soup4)
scholar_form=soup4.find_all('table',class_="gsc_mp_table")
rank=[]
for i in scholar_form:
    Rank=i.find_all('td',class_="gsc_mvt_p")
    for a in Rank:
        rank.append(a.text)
#print(rank)
```

```python
Publication=[]
for b in scholar_form:
    publication=b.find_all('td',class_="gsc_mvt_t")
    for c in publication:
        Publication.append(c.text)
#print(Publication)    
```

```python
H5_index=[]
for d in scholar_form:
    h5_index=d.find_all("td",class_="gsc_mvt_n")
    #print(h5_index)
    for x in h5_index:
        h6_index=x.find_all('a',class_="gs_ibl gsc_mp_anchor")
        for k in h6_index:
            H5_index.append(k.text)
#print(H5_index)
```

```python
H5_median=[]
for y in scholar_form:
    h5_median=y.find_all('td',class_="gsc_mvt_n")
    for z in h5_median:
        h6_median=z.find_all('span',class_="gs_ibl gsc_mp_anchor")
        for j in h6_median:
            H5_median.append(j.text)
#print(H5_median)
print(len(H5_median))
```

```python
Google_form= pd.DataFrame({})
Google_form['Rank']=rank
Google_form['Publication']=Publication
Google_form['H5_index']=H5_index
Google_form['H5_median']=H5_median
Google_form
```
